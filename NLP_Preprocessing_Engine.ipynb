#Task 1: Conceptual Understanding

1.What is the difference between "Love" and "love" in NLP?

Solution:
In Natural Language Processing, "Love" and "love" are treated as two different words
because NLP models are usually case-sensitive.

This means uppercase and lowercase letters are considered different tokens.

For example, if we do not convert text to lowercase, the model may treat "Love" and "love" as separate words,
which increases unnecessary vocabulary size and creates inconsistency.

That is why lowercasing is an important preprocessing step,
as it helps in standardizing the text and improving model performance.



2.What happens if stopwords are not removed?

Solution:
Stopwords are common words like "is", "the", "and", "in" which usually do not carry significant meaning.

If stopwords are not removed:

The dataset becomes noisy and larger in size.

Important words may get less importance
Model training may become slower and less efficient.

It may reduce the accuracy of NLP models.



3.Provide two real-world scenarios where removing stopwords can be harmful.

Solution:
There are situations where stopwords are actually important for meaning:

1.Sentiment Analysis  
Example: "I am not happy"  
Removing "not" changes the meaning completely.

2.Chatbots or Question Answering  
Example: "What is your name?"  
Removing words like "what" or "is" makes the sentence unclear.

In such cases, stopwords are important and should not be removed.



4.Explain the difference between stemming and lemmatization.

Solution:
Both stemming and lemmatization are used to reduce words to their base form, but they work differently.

Stemming:
It removes suffixes in a simple and fast way
It may not produce real words

Example:

"running" → "runn"

"better" → "bett"

Lemmatization:
It converts words into their actual root form (dictionary word)
It uses linguistic rules and is more accurate.

Example:

"running" → "run"

"better" → "good"

In short, stemming is faster but less accurate, while lemmatization is slower but more meaningful.


```python
# Task 2:Build Advanced Preprocessing Function


import re
import string

def preprocess_text(text):

    # 1. handle invalid input
    if not text or not isinstance(text, str):
        return [], ""

    # 2. lowercase
    text = text.lower()

    # 3. remove URLs
    text = re.sub(r'http\S+|www\S+', '', text)

    # 4. remove emails
    text = re.sub(r'\S+@\S+', '', text)

    # 5. remove numbers
    text = re.sub(r'\d+', '', text)

    #  remove emojis / non-ascii characters
    text = re.sub(r'[^\x00-\x7F]+', '', text)

    # 6. handle repeated characters (FIXED)
    text = re.sub(r'(.)\1{2,}', r'\1\1', text)
    text = re.sub(r'\bsoo+\b', 'so', text)
    text = re.sub(r'\bgoo+d\b', 'good', text)

    # 7. remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))

    # 8. remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # 9. tokenize
    tokens = text.split()

    # Remove very short tokens

    cleaned_tokens = []
    for word in tokens:
        if len(word) > 2 or word in ['no', 'not', 'so']:
            cleaned_tokens.append(word)

    cleaned_sentence = " ".join(cleaned_tokens)

    return cleaned_tokens, cleaned_sentence

#  TEST CODE (outside function)
examples = [
    "I have 2 dogs",
    "This is    good",
    "soooo goooood!!!",
    "WOW!!! This is GREAT!!!",
    "Visit http://example.com now",
    "I am not happy"
]

for text in examples:
     tokens, clean_text = preprocess_text(text)

     print("\nOriginal:", text)
     print("Tokens:", tokens)
     print("Cleaned:", clean_text)
```

# Note on Short Word Removal

According to the preprocessing rule, words with length ≤ 2 are removed to reduce noise in the text.
Due to this rule, short words like "is" and "so" are removed during token filtering because their length is 2.

As a result:

* "This is good" becomes "this good"
* "soooo goooood" becomes "good"

This behavior is expected based on the defined rule, even though these words may carry some meaning in the sentence.

* But "so" kept as exception.


```python
# Task 3: Stress Testing

test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

print("----- Stress Testing Results -----\n")

for i, text in enumerate(test_sentences, 1):
    tokens, clean_text = preprocess_text(text)

    print(f"Sentence {i}:")
    print("Original Text   :", text)
    print("Cleaned Tokens  :", tokens)
    print("Cleaned Sentence:", clean_text)
    print("-" * 50)
```

```python
# Task 4: Token Analytics

print("----- Token Analytics -----\n")

results = []

for i, text in enumerate(test_sentences, 1):
    tokens, clean_text = preprocess_text(text)

    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))

    # avoid division by zero
    if total_tokens > 0:
        avg_length = sum(len(word) for word in tokens) / total_tokens
    else:
        avg_length = 0

    results.append({
        "sentence_no": i,
        "original": text,
        "tokens": tokens,
        "total": total_tokens,
        "unique": unique_tokens,
        "avg_len": avg_length
    })

    print(f"Sentence {i}:")
    print("Tokens:", tokens)
    print("Total Tokens:", total_tokens)
    print("Unique Tokens:", unique_tokens)
    print("Average Token Length:", round(avg_length, 2))
    print("-" * 50)
```

```python
# Task 5: Frequency Analysis

from collections import Counter

# Combine all tokens from all sentences
all_tokens = []

for text in test_sentences:
    tokens, _ = preprocess_text(text)
    all_tokens.extend(tokens)

# Create frequency counter
word_freq = Counter(all_tokens)

print("----- Frequency Analysis -----\n")

# Top 10 most frequent words
print("Top 10 Most Frequent Words:")
top_10 = word_freq.most_common(10)
for word, count in top_10:
    print(f"{word} : {count}")

print("\n" + "-"*40 + "\n")

# Top 5 least frequent words
print("Top 5 Least Frequent Words:")

# words with lowest frequency (sorted)
least_common = sorted(word_freq.items(), key=lambda x: x[1])[:5]

for word, count in least_common:
    print(f"{word} : {count}")
```

```python
# Task 6: Build Full Pipeline

def full_pipeline(text_list):

    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, clean_text = preprocess_text(text)

        all_tokens.append(tokens)
        clean_sentences.append(clean_text)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
result = full_pipeline(test_sentences)

print("Tokens:\n", result["tokens"])
print("\nClean Sentences:\n", result["clean_sentences"])
```
