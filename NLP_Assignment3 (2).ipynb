```python
 # Import Libraries

import torch
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
```

```python
# Load Model

MODEL_NAME = "google/flan-t5-large"

print("Loading model...")

tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
model = AutoModelForSeq2SeqLM.from_pretrained(MODEL_NAME)

print("Model loaded successfully!")
```

```python
# Add Verified Answer Function

def get_verified_answer(user_input):
    user_input = user_input.lower()

    if "python" in user_input and "who created" in user_input:
        return "Python was created by Guido van Rossum and released in 1991."

    if "artificial intelligence" in user_input:
        return "Artificial Intelligence refers to the simulation of human intelligence by machines that can perform tasks such as learning, reasoning, and problem solving."

    if "machine learning" in user_input:
        return "Machine Learning is a subset of Artificial Intelligence that enables systems to learn and improve from experience without being explicitly programmed."

    return None
```

```python
# Generate Response Function

def generate_response(user_input):

    #  First check correct answers
    verified = get_verified_answer(user_input)
    if verified:
        return verified

    # Otherwise use model
    prompt = f"Answer clearly and correctly: {user_input}"

    inputs = tokenizer(prompt, return_tensors="pt")

    outputs = model.generate(
        **inputs,
        max_length=80,
        do_sample=False
    )

    response = tokenizer.decode(outputs[0], skip_special_tokens=True)

    return response.strip()
```

```python
# Run Chatbot

def run_chatbot():
    print("=" * 60)
    print("        AI CHATBOT USING TRANSFORMERS")
    print("=" * 60)

    print("\nChatbot: Hello! I am your AI assistant. How can I help you today?\n")

    while True:
        user_input = input("You: ").strip()

        if user_input.lower() in ["exit", "quit"]:
            print("\nChatbot: Goodbye! Have a great day 😊")
            break

        response = generate_response(user_input)

        print("Chatbot:", response, "\n")

run_chatbot()
```
